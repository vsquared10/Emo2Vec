```python
em_words['word'], em_words['synonym'] = em_words['word'].str.split('--').str
```

```python
em_words.head()
```

```python
em_words['toss1'], em_words['joy'] = em_words['joy'].str.split('-').str
em_words['toss2'], em_words['sadness'] = em_words['sadness'].str.split('-').str
em_words['toss3'], em_words['fear'] = em_words['fear'].str.split('-').str
em_words['toss4'], em_words['anger'] = em_words['anger'].str.split('-').str
em_words['toss5'], em_words['trust'] = em_words['trust'].str.split('-').str
em_words['toss6'], em_words['disgust'] = em_words['disgust'].str.split('-').str
em_words['toss7'], em_words['surprise'] = em_words['surprise'].str.split('-').str
em_words['toss8'], em_words['anticipation'] = em_words['anticipation'].str.split('-').str
em_words['toss9'], em_words['POS'] = em_words['POS'].str.split('-').str
em_words.head()
```

```python
em_words.drop(['toss1','toss2','toss3','toss4','toss5','toss6','toss7','toss8','toss9'],
              axis=1,
              inplace=True)
```

```python
em_words.head()
```

```python
new_cols = ['annotator_id',
                    'word','synonym',
                    'joy',
                    'sadness',
                    'fear',
                    'anger',
                    'trust',
                    'disgust',
                    'surprise',
                    'anticipation',
                    'POS']
em_words = em_words.reindex_axis(new_cols, axis=1)
em_words.head()
```

```python
em_words.tail()
```

```python
o_size=len(em_words)
o_size
```

```python
emotions = em_words[['joy',
                     'sadness',
                     'fear',
                     'anger',
                     'trust',
                     'disgust',
                     'surprise',
                     'anticipation']]
emotions.head()
```

```python
em_words[emotions.columns] = em_words[emotions.columns].apply(pd.to_numeric)
type(em_words['joy'][0])
```

```python
affected = em_words[emotions.columns].groupby([em_words['word']], sort=False).mean().reset_index()
```

```python
def hists():    
    sns.set(style="white", palette="muted", color_codes=True)
    for emo in emotions.columns:
        sns.distplot(affected[emo], label=emo, axlabel='Average Score')
    plt.title('Word Emotion Score')
    plt.legend()
    return plt.tight_layout();
```

```python
# Load Google's pre-trained Word2Vec model.
model = gensim.models.Word2Vec.load_word2vec_format('https://s3-us-west-1.amazonaws.com/victorsdatasets/GoogleNews-vectors-negative300.bin', binary=True)  
```

```python
# built this as some of the words originally didn't exist in google's pre-trained model
counter = 0
for word in affected['word']:
    if word in model.vocab:
        counter += 1
print(counter, "words or", np.round(counter/len(affected)*100.0, 2), "% of the words are in google's set")
```

```python
# create list of word indicies to drop to avoid keyerrors with Google's pre-trained model.
idx_to_drop = []
def dropper():
    for ex in affected['word']:
        if ex not in model.vocab:
            idx_to_drop.append(affected.loc[affected.word == ex].index[0])     
```

```python
dropper()
affected = affected.drop(idx_to_drop, axis=0)
```

```python
G_vectors = lambda x: model[x]
affected['word_vectors'] = affected['word'].apply(G_vectors)
```

```python
affected.head()
```

```python
affected['label_vectors'] = affected[emotions.columns].values.tolist()
```

```python
affected.head()
```

```python
affected['binary'] = (affected[emotions.columns] > 0).astype(int).values.tolist()
```

```python
affected.head()
```

```python
df1 = affected[emotions.columns].rank(method='max', axis=1).rank(method='first', axis=1)
ma = df1.max().max()
```

```python
affected['label'] = (df1== ma).astype(int).values.tolist()
```

```python
affected['target'] = affected['label'].apply(lambda x: x.index(1))
```

```python
label_dict = {0 : 'joy', 1 : 'sadness', 2 : 'fear', 3 : 'anger', 4 : 'trust', 5 : 'disgust', 6 : 'surprise', 7 : 'anticipation'}
```

```python
affected['label_name'] = affected['target'].apply(lambda x: label_dict[x])
```

```python
emo2vec = affected[['word_vectors', 'label_vectors', 'binary', 'label', 'target', 'label_name']]
emo2vec.head()
```

# Model Testing

```python
emo_X, emo_y = list(emo2vec.word_vectors), list(emo2vec.target)
emo_X_train, emo_X_test, emo_y_train, emo_y_test = train_test_split(emo_X, emo_y, random_state=42)
```

```python
# code from an early ML lab
def plot_embedding(X, y, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure(figsize=(10, 6), dpi=250)
    ax = plt.subplot(111)
    ax.axis('off')
    ax.patch.set_visible(False)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], str(y[i]), color=plt.cm.Set1(y[i] / 10.), fontdict={'weight': 'bold', 'size': 12})

    plt.xticks([]), plt.yticks([])
    plt.ylim([-0.1,1.1])
    plt.xlim([-0.1,1.1])

    if title is not None:
        plt.title(title, fontsize=16)
```

```python
pca = PCA(n_components=2)
plot_X = pca.fit_transform(normed_scaled)
plot_y = emo_y
plot_embedding(plot_X, plot_y)
```

### Attempt to fit kmeans

```python
def plot_centroids(data, labels, n):
    '''
    Input:
        data - (DataFrame) Data to cluster on
        labels - (list) List containing the labels for each datapoint
    Output:
        Plot showing the original datapoints and cluster centers found by KMeans
    '''
    km = KMeans(n_clusters=n).fit(data)
    cluster_x = [km.cluster_centers_[i][0] for i in range(len(km.cluster_centers_))]
    cluster_y = [km.cluster_centers_[i][1] for i in range(len(km.cluster_centers_))]
    plt.scatter(data[:, 0], data[:, 1], color='blue', alpha=.1)
    plt.scatter(cluster_x, cluster_y, color='red')
    plt.show()
```

```python
KM = KMeans(n_clusters=8, n_jobs=-1)
KM.fit(plot_X)
plot_centroids(plot_X, KM.labels_, 8)
```

```python
# from 6003 lab on GMM
def plot_silhouette(model_library,data,components):
    '''
    Input:
        data - (DataFrame) Data to cluster on
        clusters - (list) List containing the number of clusters to check for
    Output:
        Plot showing the silhouette score for different numbers of centroids
    '''  
    labels = []
    silhouettes = []
    if model_library == GaussianMixture:
        for c in components:
            labels.append(model_library(n_components=c).fit(data).predict(data))
    elif model_library == KMeans:
        for c in components:
            labels.append(model_library(n_clusters=c).fit(data).predict(data))
    else:
        pass
    for label in labels:
        silhouettes.append(silhouette_score(data, label, metric='euclidean'))
    enumerated = dict(enumerate(silhouettes))
    return plt.plot(components, silhouettes), enumerated
```

```python
new_tests = range(2, 10)
plot, sils = plot_silhouette(GaussianMixture,plot_X, new_tests)
plot
print("{0} clusters result(s) in highest Silhouette Score of {1}.".format(max(sils, key=lambda k: sils[k])+new_tests[0], sils[max(sils, key=lambda k: sils[k])])) 
```

```python
def plot_means_gmm(data, n):
    '''
    Input:
        data - (DataFrame) Data to cluster on
        labels - (list) List containing the labels for each datapoint
    Output:
        Plot showing the original datapoints and cluster centers found by KMeans
    '''
    gmm = GaussianMixture(n_components=n).fit(data)
    mean_x = [gmm.means_[i][0] for i in range(len(gmm.means_))]
    mean_y = [gmm.means_[i][1] for i in range(len(gmm.means_))]
    plt.scatter(data[:, 0], data[:, 1], color='blue', alpha=.1)
    plt.scatter(mean_x, mean_y, color='red')
    plt.show()
```

```python
plot_means_gmm(plot_X, 3)
```
